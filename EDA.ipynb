```python
import pandas as pd
rent = pd.read_csv("DataRent.csv")
sold = pd.read_csv("DataSold.csv")
```

# Type of houses

```python
rent['TYPE'].value_counts()
```

```python
import matplotlib.pyplot as plt

values = rent['TYPE'].value_counts()

ax = values.plot(kind="bar",figsize=(10,4),color="orange")
plt.title("Type of property counts (rental)")

for index, value in enumerate(values):
    plt.text(index, value + 0.5, str(value), ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt

values = sold['TYPE'].value_counts()

ax = values.plot(kind="bar",figsize=(11,4),color="green")
plt.title("Type of property counts (for sale)")

for index, value in enumerate(values):
    plt.text(index, value + 0.5, str(value), ha='center', va='bottom')

plt.show()
```
